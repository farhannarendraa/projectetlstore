Program ini bertujuan cek validitas data yang sudah dilakukan transformasi di proses ETL pada data The Global Superstore menggunakan Great Expectation.



```python
# Import Library
import great_expectations as gx
import pandas as pd
from great_expectations.data_context import FileDataContext
```

```python
df = pd.read_csv(
    "project-m3/dags/P2M3_Farhan_Narendra_Achyara_data_clean.csv"
)
df
```

```python
# Create Data Context

context = gx.get_context()
```

```python
# Create a data context

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'milestone3'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'milestone3_data'
path_to_data = 'project-m3/dags/P2M3_Farhan_Narendra_Achyara_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation_global_superstore'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

```python
# Ekspektasi 1: order_id tidak boleh ada missing value

result1 = validator.expect_column_values_to_not_be_null('order_id')
```

Kolom order_id tidak ada missing value. Karena jika ada missing value, akan menyebabkan data sulit diidentifikasi karena tidak adanya unique identifier.

```python
# Ekspektasi 2: column quantity berada di antara nilai minimum dan maksimum

result2 = validator.expect_column_values_to_be_between(
    column="quantity",
    min_value=1,
    max_value=20,
    mostly=0.95  # Allow 5% outliers
)
```

Penggunaan ekspektasi ini dimaksudkan untuk melihat apakah ada anomali pada data.

```python
# Ekspektasi 3: column segmen sesuai dengan kategorinya berupa consumer, corporate, dan home office

valid_segments = ["Consumer", "Corporate", "Home Office"]
result3 = validator.expect_column_values_to_be_in_set(
    column="segment",
    value_set=valid_segments
)
```

Penggunaan ekspektasi ini dimaksudkan untuk melihat apakah segmen produk perusahaan sesuai pada tempatnya untuk kolom segment.

```python
# Ekspektasi 4: kolom profit sesuai dengan tipenya yaitu numerik

result4 = validator.expect_column_values_to_be_in_type_list(
    column="profit",
    type_list=["int64", "int32", "int", "float64", "float"]
)
```

Penggunaan ekspektasi ini dimaksudkan agar seluruh data dalam kolom profit merupakan numerik seluruhnya sehingga mempermudah visualisasi data.

```python
# Ekspektasi 5: kolom A yaitu sales lebih besar daripada kolom B yaitu shipping cost

result5 = validator.expect_column_pair_values_a_to_be_greater_than_b(
    column_A="sales",
    column_B="shipping_cost",
    or_equal=True
)
```

Penggunaan ekspektasi ini dimaksudkan untuk melihat apakah harga sales melebihi harga shipping karena jika harga shipping lebih tinggi akan menyebabkan kerugian pada perusahaan.

```python
# Ekspektasi 6: kolom customer_name memiliki regular expression yang sama

result6 = validator.expect_column_values_to_match_regex(
    column="customer_name",
    regex="^a.*",
)
```

Penggunaan ekspektasi ini dimaksudkan agar nama pelanggan sesuai dengan nama aslinya untuk mempermudah pengiriman barang.

```python
# Ekspektasi 7: kolom sales memiliki nilai rata-rata di antara nilai minimum dan maksimum

result7 = validator.expect_column_mean_to_be_between(
    column = "sales",
    min_value = 500,
    max_value = 5000
)
```

Penggunaan ekspektasi ini dimaksudkan untuk melihat anomali pada harga rata-rata barang yang dijual.

Seluruh data dinyatakan valid berdasarkan 7 ekspektasi dari Great Expectation.
